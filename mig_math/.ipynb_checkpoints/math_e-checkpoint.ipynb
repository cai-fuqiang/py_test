```python
import plotly.graph_objs as go
import numpy as np
import pandas as pd
import re

def get_trace(_x, _y, _trace_name):
    trace = go.Scatter(
        x=_x,  
        y=_y, 
        # 散点 + 线段绘图
        mode="markers + lines",
        name=_trace_name,
    )
    return trace
def get_figure(_x_name, _y_name):
    fig = go.Figure(
        layout={
            "template": "xgridoff",
            "height": 500,
            "xaxis_title": _x_name,
            "yaxis_title": _y_name,
            "xaxis_tickformat": ',',
            "yaxis_tickformat": ',',
        }
    )
    return fig
def add_trace_to_figure(trace, fig):
    fig.add_trace(trace)

def show_fig(fig):
    fig.show()

def do_one_iter(bw, remain_mem, dirtyrate, delta):
    new_remain_mem = remain_mem + dirtyrate  * (delta / 1000)
    return new_remain_mem

def get_total_time_all(bw: float, total_mem: float, downtime: float, dirtyrate: float):
    remain_mem = total_mem
    iter_time = 0
    delta = 0
    total_time = 0
    while True:
        if downtime * bw / 1000 > remain_mem:
            remain_mem=do_one_iter(bw, remain_mem, dirtyrate, delta)
            iter_time +=1
            #print("iter_time:", iter_time, "remaining:", remain_mem, "total_time:", total_time)
            
            if downtime * bw / 1000 > remain_mem:
                break
            delta=0
        delta += 100
        total_time += 100
        ## print(delta)
        remain_mem -=  bw* 100 / 1000;

    return total_time
```

## 不同脏页速率影响 

```python

def get_dirty_rate_arr(bw):
    dirty_rate=50
    dirty_rate_arr = []
    while dirty_rate < bw:
        dirty_rate_arr.append(dirty_rate)
        total_time_arr.append(total_time)
        dirty_rate=dirty_rate+10
    return dirty_rate_arr

def get_total_time(dirty_rate_arr, total_mem, bw, downtime):
    total_time_arr = []
    for dirty_rate in dirty_rate_arr:
        total_time = get_total_time_all(bw, total_mem, downtime, dirty_rate)
        total_time_arr.append(total_time)
    return total_time_arr

bw = 90 * 1024 / 8  # 11.25 GB/s
bw = 11760

total_mem = 320 * 1024
fig = get_figure("dirtyrate", "totaltime")
dirty_rate_arr = get_dirty_rate_arr(bw)


total_time_arr = get_total_time(dirty_rate_arr, total_mem, bw, 300)
trace = get_trace(dirty_rate_arr, total_time_arr, "m_320_bw_90G_down_300")
add_trace_to_figure(trace, fig)

total_mem = 160 * 1024
total_time_arr = get_total_time(dirty_rate_arr, total_mem, bw, 300)
trace = get_trace(dirty_rate_arr, total_time_arr, "m_160_bw_90G_down_300")
add_trace_to_figure(trace, fig)

total_mem = 80 * 1024
total_time_arr = get_total_time(dirty_rate_arr, total_mem, bw, 300)
trace = get_trace(dirty_rate_arr, total_time_arr, "m_80_bw_90G_down_300")
add_trace_to_figure(trace, fig)

total_mem = 40 * 1024
total_time_arr = get_total_time(dirty_rate_arr, total_mem, bw, 300)
trace = get_trace(dirty_rate_arr, total_time_arr, "m_40_bw_90G_down_300")
add_trace_to_figure(trace, fig)

show_fig(fig)
```
